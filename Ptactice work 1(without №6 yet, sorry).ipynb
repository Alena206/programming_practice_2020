```python
 #Упражнение 5
import numpy as np
import matplotlib.pyplot
x = [1, 2, 3, 4, 5, 6]
y = [2.43, 2.55, 2.67, 2.88, 3.16, 3.42]
p, v = np.polyfit(x, y, deg=1, cov=True)
p_f = np.poly1d(p)
x1 = np.arange(1, 6, 0.0001)
plt.plot(x,y,x1,p_f(x1))
plt.show()
```
